## Preparation for Modelling 

- From the EDA, the classes in this situation seemed to be imbalanced. So in order to handle the imbalance of classes, the approach to address these would be the implementation of the Synthetic Minority Oversampling Technique **(SMOTE)** which is a commonly used class balancing technique. 

- Also, SMOTE is an oversampling technique where the synthetic samples are generated for the minority class and this algorithm helps to overcome the overfitting problem posed by random oversampling. 

-  We will proceed the modelling process by using the SMOTE with 4 different modelling evaluation namely Logistic Regression, Random Forest, XGBoost Classifier and GradientBoost Classifier.

## Custom Functions
---

- Function calls for computing modelling scores

```python
"""
Function call to assign a selected model and its parameters to compute the model scores such as
train score, test score, accuracy, sensitivity, specificity, precision, f1score
"""
def model(model, params, name):
    """
    Inputs: Model, Parameters for the Model, Name of the Model
    Output: Runs the corresponding model and saves scores to results dictionary
    """
    gs = GridSearchCV(model,params,cv=3,verbose=1,n_jobs=-1, scoring='roc_auc') # Runs Model 
    gs.fit(X_train_sm, y_train_sm)
    print(f'Best Cross-Validation Score: {round(gs.best_score_, 3)}')
    print(f'ROC-AUC Score on Train Set: {round(gs.score(X_train_sm, y_train_sm), 3)}')
    print(f'ROC-AUC Score on Test Set: {round(gs.score(X_test_sc, y_test), 3)}')
    print(gs.best_params_)

    print('')
    
    pred = gs.predict(X_test_sc) # Generates Confusion Matrix and Scores
    print(pd.DataFrame(confusion_matrix(y_test, pred),
                         columns=['pred_nownv', 'pred_wnv'],
                         index=['actual_nownv', 'actual_wnv']))
    tn, fp, fn, tp = confusion_matrix(y_test, pred).ravel()
    accuracy = (tn + tp) / (tn + fp + fn + tp)
    print(f'Accuracy: {round(accuracy, 3)}')
    sensitivity = tp / (tp + fn)
    print(f'Sensitivity: {round(sensitivity, 3)}')
    specificity = tn / (tn + fp)
    print(f'Specificity: {round(specificity, 3)}')
    precision = tp / (tp + fp)
    print(f'Precision: {round(precision, 3)}')
    f1score = f1_score(y_test, pred)
    print(f'F1 Score: {round(f1score, 3)}')

    results[name] = [gs.score(X_train_sm, y_train_sm), gs.score(X_test_sc, y_test), accuracy, sensitivity, specificity, precision, f1score] # Saves scores to result dictionary
    return gs
```

```python
# Function call to store into a csv file for kaggle submission with a computed model to be provided
def kaggle(model, csvname):
    """
    Input: Model, Name of CSV
    Outputs : Kaggle Submission saved as model.csv
    """
    submission = pd.DataFrame()
    pred = pd.DataFrame(model.predict_proba(test_sc)[:,1])
    submission = pd.concat([test_id, pred], axis=1)
    submission.columns =['Id', 'WnvPresent']
    submission.to_csv(f'../predictions/{csvname}.csv', index=False)
    print('Submission created')
```

## Pre-processing for Modelling
---

```python
train = pd.read_csv('../assets/train_clean_model.csv')
test_1 = pd.read_csv('../assets/test1_clean_model.csv')
test_2 = pd.read_csv('../assets/test2_clean_model.csv')
test = pd.concat([test_1, test_2]).reset_index(drop=True)
```

```python
# Saving test id data for joining later for kaggle submission
test_id = test['id'].copy()
```

```python
train.drop(columns=['date', 'trap', 'station'], inplace=True)
test.drop(columns=['date', 'trap', 'station', 'species-unspecified_culex', 'id'], inplace=True)
```

```python
# Dummifying cluster variable
train = pd.get_dummies(train, columns=['cluster'], prefix_sep='-')
test = pd.get_dummies(test, columns=['cluster'], prefix_sep='-')
train.drop(columns='cluster-0', inplace=True)
test.drop(columns='cluster-0', inplace=True)
```

```python
# Creating features and target 
X = train.drop(columns=['nummosquitos', 'wnvpresent'])
y = train['wnvpresent']
```

```python
# Checking value counts
y.value_counts(normalize=True)
```

```python
# split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.25, stratify=y)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Modelling & Evaluation
---

The modelling workflow will be as such: <br/>
    1. Using `StandardScaler()` to scale the data<br/>
    2. Using an estimator of choice (e.g. Logistic Regression, Random Forest, XGB Classifier)<br/>
    3. Using GridSearchCV to find the best hyperparameters for fine-tuning the model.<br/>
    4. Evaluate the model using AUC score on the test set

```python
ss = StandardScaler()
X_train_sc = ss.fit_transform(X_train)
X_test_sc = ss.transform(X_test)
```

```python
test_sc = ss.transform(test)
```

```python
# Using SMOTE to bootstrap data to create new training data
oversample = SMOTE(random_state=42)
X_train_sm, y_train_sm = oversample.fit_resample(X_train_sc, y_train)
```

```python
X_train_sm = pd.DataFrame(X_train_sm, columns=X.columns.tolist())
X_test_sc = pd.DataFrame(X_test_sc, columns=X.columns.tolist())
```

### Setting up Models

```python
results = {}
```

### Results

```python
df_results = pd.DataFrame(results, index=['train_set_auc_score', 'test_set_auc_score', 'accuracy', 'sensitivity', 'specificity', 'precision', 'f1_score']).T
df_results = round(df_results,3)
```

```python
df_results.sort_values('sensitivity')
```

In predicting the occurrence of West Nile Virus, we assume that the cost from false negatives i.e. not detecting the virus resulting in an outbreak is far higher than the cost of false positives i.e spraying the location. Thus as we are concerned with minimizing the number of false negatives, senstivity was chosen as the key metric to evaluate the models by.

The best model in this regard was the Logistic Regression model with a high sensitivity score on the test set. This may be the case as the explanatory variables had many multicollineared features that resulted in large variance in the other models, which was regularised by logistic regression. Nevertheless, the logistic regression model was. not perfect - its sensitivity came at some expense of increased false positives, with a low precision and f1 score as a result.
